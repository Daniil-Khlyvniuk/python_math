# 4-Bit Binary Adder

Screenshot 2024-09-19 at 15.11.16.png


```python
input_a = [ True, False ] # 2 in 10; A0 - "1"; A1 - "0";
input_b = [ True, True ] # 3 in 10; B0 - "1"; B1 - "1";
```

```python
def XOR(a, b):
  return (a and not b) or (not a and b) # or just (a ^ b) in python syntaxis
```

```python
def half_adder(a, b):
  S = XOR(a, b) # Sum of "a" and "b"
  C_out = a and b # Curry bit

  return S, C_out
```

```python
def full_adder(a, b, c_in):
  S_prev, C = half_adder(a, b)
  S_current = XOR(S_prev, c_in)
  C_out = (S_current and c_in) or (C)

  return S_current, C_out
```

```python
def binary_add(a, b):
  S_1, C_in = half_adder(a[-1], b[-1])
  S_2, S_3 = full_adder(a[-2], b[-2], C_in)

  return [ False, S_3, S_2, S_1 ]
```

```python
print(''.join(str(int(n)) for n in input_a) + ' + ' + ''.join(str(int(n)) for n in input_b) + ' = ' + ''.join(str(int(n)) for n in binary_add(input_a, input_b)))
```
