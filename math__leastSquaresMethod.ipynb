# Least squares method

The least squares method is a statistical procedure to find the best fit for a set of data points. The method works by minimizing the sum of the offsets or residuals of points from the plotted curve. Least squares regression is used to predict the behavior of dependent variables.

[Read more about least squares method](https://en.wikipedia.org/wiki/Least_squares)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 300      # count of experiments
sigma = 3    # standard deviation of observed values
k = 0.5      # theoretical value of param "k"
b = 2        # theoretical value of param "b"
```

```python
x = np.array(range(N))
function = np.array([ k * z + b for z in range(N) ])
experimental_values_Y = function + np.random.normal(0, sigma, N)
```

```python
plt.plot(function, c='blue')
plt.scatter(x, experimental_values_Y, s=2, c='red')
plt.grid = True
plt.show
```

```python
mx = x.sum() / N
my = experimental_values_Y.sum() / N
alfa_2 = (x.T @ x) / N
alfa_11 = (x.T @ experimental_values_Y) / N
```

```python
kk = (alfa_11 - mx*my)/(alfa_2 - mx**2)
bb = my - kk*mx
ff = np.array([kk*z+bb for z in range(N)])


print('b =', bb)
print('k =', kk)
```

```python
plt.scatter(x, experimental_values_Y, s=2, c='red')
plt.plot(ff, c='green')
plt.show()
```
